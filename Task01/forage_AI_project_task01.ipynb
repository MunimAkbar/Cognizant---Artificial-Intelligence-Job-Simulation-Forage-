```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(y=data.category)
for i in ax.containers:
  ax.bar_label(i)
plt.title(f'Count of product solds from each {len(data.category.unique())} categories')
plt.show()
```

```python

# Distribution of unit prices
sns.histplot(data['unit_price'], kde=True)
plt.title('Distribution of Unit Prices')
plt.show()

# Distribution of quantities
sns.histplot(data['quantity'], kde=True)
plt.title('Distribution of Quantities')
plt.show()

# Distribution of total sales
sns.histplot(data['total'], kde=True)
plt.title('Distribution of Total Sales')
plt.show()

# Relationship between unit price and total sales
sns.scatterplot(data=data, x='unit_price', y='total')
plt.title('Unit Price vs. Total Sales')
plt.show()

# Relationship between quantity and total sales
sns.scatterplot(data=data, x='quantity', y='total')
plt.title('Quantity vs. Total Sales')
plt.show()
```

```python
# Sample data to replicate the issue
data = pd.DataFrame({
    'timestamp': pd.date_range(start='2022-01-01', periods=100, freq='D'),
    'total': range(100)
})

# Convert the timestamp column to datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])

# Set the timestamp as the index
data.set_index('timestamp', inplace=True)

# Inspect the data
print(data.head())
```

```python
# Plot total sales over time
data['total'].resample('M').sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
